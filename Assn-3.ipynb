### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )
plt.plot(a, 'ko-')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
b   = np.load( 'data-fig2.npy' )

plt.plot(b[:,0], 'r', label='Red')
plt.plot(b[:,1], 'g', label='Green')
plt.plot(b[:,2], 'b', label='Blue')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c   = np.load( 'data-fig3.npy' )

plt.figure()
plt.scatter(c[:,0], c[:,1], color='b', alpha=0.3, s=200)

plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d   = np.load( 'data-fig4.npy' )
x = ['A', 'B', 'C', 'D', 'E']
y = d

plt.figure()
plt.bar(x, y, color='gray', alpha=0.6)
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e   = np.load( 'data-fig5.npy' )

plt.figure()
plt.imshow(e, cmap='RdBu', origin='lower', aspect='auto', vmin=-1, vmax=1)
plt.colorbar()
plt.show()
```
